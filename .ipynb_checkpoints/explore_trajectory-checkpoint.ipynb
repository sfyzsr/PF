```python
path = '/data/processed/peize/sfrs_data/'
# path = '/data/processed/peize/waverley_data/'
paths = []
folders = os.listdir(path)
folders.sort()
for folder in folders:
    if 'Pro 1' in folder:
        paths.append(path + folder + '/')
```

```python
for path in paths:

    try:
        trajectory = pd.read_csv(path + 'trajectory.csv')
    except:
        print("File not found. Skipping trajectory: ", path + 'trajectory.csv')
        continue
    timestamps = pd.read_csv(path + 'saved_timestamps.csv')
    
    xs = trajectory['x_right[m]'].tolist()
    ys = trajectory['y_right[m]'].tolist()

    ts = [t/1000 for t in timestamps['timestamp_global[ms]']]

   
    ts_ixs_tr = np.searchsorted(trajectory['t[s]'], ts)
    print(path)
    print(ts_ixs_tr.shape)

    plt.plot(xs, ys)
    try:
        plt.plot([xs[i] for i in ts_ixs_tr], [ys[i] for i in ts_ixs_tr], 'o', markersize = 10)
    except:
        plt.axis('equal')
        plt.show()
        continue
    plt.axis('equal')
    plt.show()
```

```python
for path in paths:

    try:
        trajectory = pd.read_csv(path + 'trajectory_ncl_soft.csv')
    except:
        print("File not found. Skipping trajectory: ", path + 'trajectory.csv')
        continue
    timestamps = pd.read_csv(path + 'saved_timestamps.csv')
    
    xs = trajectory['x_avg[m]'].tolist()
    ys = trajectory['y_avg[m]'].tolist()

    ts = [t/1000 for t in timestamps['timestamp_global[ms]']]

   
    ts_ixs_tr = np.searchsorted(trajectory['t[s]'], ts)
    print(path)
    print(ts_ixs_tr.shape)

    plt.plot(xs, ys)
    try:
        plt.plot([xs[i] for i in ts_ixs_tr], [ys[i] for i in ts_ixs_tr], 'o', markersize = 10)
    except:
        plt.axis('equal')
        plt.show()
        continue
    plt.axis('equal')
    plt.show()
```

```python
for path in paths:

    magnetic = pd.read_csv(path + 'magnetic_field_5.csv', skiprows=[0])

    mx = [x for x in magnetic['mfield_x[uT]']]
    my = [x for x in magnetic['mfield_y[uT]']]
    mz = [x for x in magnetic['mfield_z[uT]']]

    plt.plot(mx, label="x")
    plt.plot(my, label="y")
    plt.plot(mz, label="z")

    if max(mx) < 200 and max(my) < 200 and max(mz) < 200:
        plt.ylim(-100, 100)
    plt.legend()
    plt.show()
```
