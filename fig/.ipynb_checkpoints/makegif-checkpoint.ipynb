```python
import os
import imageio
fig = 'online23'
fig_dir = './' + fig
def get_number(s):
    return int(s.split('_')[0].split('\\')[-1])
# List all the figure files in the directory
fig_files = sorted([os.path.join(fig_dir, f) for f in os.listdir(fig_dir)],key=get_number)

# Load the figure files into a list of images
images = [imageio.imread(f) for f in fig_files]

# Save the list of images as a GIF animation
imageio.mimsave(fig+'.gif', images, fps=5)
```
